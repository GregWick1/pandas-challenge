# PyCity Schools Analysis

```python
# Import dependencies
import pandas as pd
from pathlib import Path
# Load csv files
school_data_load = Path('Resources/schools_complete.csv')
student_data_load = Path('Resources/students_complete.csv')
# read the files into pandas                    
school_data = pd.read_csv(school_data_load)
student_data = pd.read_csv(student_data_load)
```

```python
school_data.head()
```

```python
student_data.head()
```

```python
# Merge the two files into one dataset
school_data_complete = pd.merge(student_data, school_data, how='left', on=['school_name', 'school_name'])
school_data_complete.head()
```

## District Summary

```python
# Total number of unique schools
school_count = len(school_data_complete['school_name'].unique())
school_count
```

```python
# Total number of students
student_count = school_data_complete['student_name'].count()
student_count
```

```python
# Total budget for schools
total_budget = school_data['budget'].sum()
total_budget
```

```python
# Average math score for all schools
average_math = school_data_complete['math_score'].mean()
average_math
```

```python
# Average reading score for all schools
average_reading = school_data_complete['reading_score'].mean()
average_reading
```

```python
# Percentage of students who passed math 
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()['student_name']
passing_math = passing_math_count / student_count * 100
passing_math
```

```python
# Percentage of students who passed reading
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()['student_name']
passing_reading = passing_reading_count / student_count * 100
passing_reading 
```

```python
# Percentage of students who are passing both math and reading
overall_passing_count = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)].count()['student_name']
overall_passing = overall_passing_count / student_count * 100
overall_passing
```

```python
# Summary DataFrame to store all the calculations we just made
school_summary = pd.DataFrame({"Total Schools": [school_count], "Total Students": student_count, "Total Budget": total_budget,
                               "Average Math Score": average_math, "Average Reading Score": average_reading, "Passing Math Score": passing_math,
                               "Passing Reading Score": passing_reading, "Overall Passing Score": overall_passing})
# Formatting changes
school_summary["Total Students"] = school_summary["Total Students"].map("{:,}".format)
school_summary["Total Budget"] = school_summary["Total Budget"].map("${:,.2f}".format)

school_summary
```

## School Summary

```python
# Setting index to school_name and making the next column school type
school_types = school_data.set_index(["school_name"])['type']
school_types.head()
```

```python
# Counting the number of students in each school
school_students = school_data_complete['school_name'].value_counts()
school_students
```

```python
# Finding the budget for each school
school_budget = school_data_complete.groupby(['school_name'])['budget'].mean()
school_budget
```

```python
# Finding the average amount of money spent on each student
school_per_capita = school_budget / school_students
school_per_capita
```

```python
# Average math score for each school
average_math_perschool = school_data_complete.groupby(['school_name'])['math_score'].mean()
average_math_perschool
```

```python
# Average reading score for each school
average_reading_perschool = school_data_complete.groupby(['school_name'])['reading_score'].mean()
average_reading_perschool
```

```python
# Percentage of students who passed math per school
passing_math_perschool = school_data_complete[(school_data_complete["math_score"] >= 70)]
passing_math_ps = passing_math_perschool.groupby(['school_name']).size()
passing_math_ps1 = passing_math_ps / school_students * 100
passing_math_ps1
```

```python
# Percentage of students who passed reading per school
passing_reading_perschool = school_data_complete[(school_data_complete["reading_score"] >= 70)]
passing_reading_ps = passing_reading_perschool.groupby(['school_name']).size()
passing_reading_ps1 = passing_reading_ps / school_students * 100
passing_reading_ps1
```

```python
# Percentage of students who passed math and reading per school
students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
school_students_passing_math_and_reading1 = school_students_passing_math_and_reading / school_students * 100
school_students_passing_math_and_reading1
```

```python
# Summary dataframe for the previous calculations
per_school_summary = pd.DataFrame({"School Type": school_types, "Total Students": school_students, "Total School Budget": school_budget,
                                   "Per Student Budget": school_per_capita, "Average Math Score": average_math_perschool, "Average Reading Score": average_reading_perschool,
                                   "Passing Math %": passing_math_ps1, "Passing Reading %": passing_reading_ps1, "Passing Overall %": school_students_passing_math_and_reading1})
# Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
per_school_summary
```

## Highest Performing Schools

```python
# Sorting the summary dataframe for the schools with the top 5 Passing Overall Percentages
top_schools = per_school_summary.sort_values(by="Passing Overall %", ascending=False)
top_schools.head()
```

## Lowest Performing Schools

```python
# Sorting the dataframe to find the schools with the bottom 5 Passing Overall Percentages
bottom_schools = per_school_summary.sort_values(by="Passing Overall %", ascending=True)
bottom_schools.head()
```

## Math Scores by Grade

```python
# Filter the data to get the data for all students seperated by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]
# Group the data by school name and then look for the average math score for each grade
ninth_graders_math = ninth_graders.groupby(["school_name"])['math_score'].mean()
tenth_graders_math = tenth_graders.groupby(["school_name"])['math_score'].mean()
eleventh_graders_math = eleventh_graders.groupby(["school_name"])['math_score'].mean()
twelfth_graders_math = twelfth_graders.groupby(["school_name"])['math_score'].mean()
# Store the data in a dataframe
math_scores_by_grade = pd.DataFrame({"9th": ninth_graders_math, "10th": tenth_graders_math, "11th": eleventh_graders_math,
                                     "12th": twelfth_graders_math})
# Formatting
math_scores_by_grade.index.name = None
math_scores_by_grade
```

## Reading Scores by Grade

```python
# Group the data by school name and find the average reading score per grade
ninth_graders_reading = ninth_graders.groupby(["school_name"])['reading_score'].mean()
tenth_graders_reading = tenth_graders.groupby(["school_name"])['reading_score'].mean()
eleventh_graders_reading = eleventh_graders.groupby(["school_name"])['reading_score'].mean()
twelfth_graders_reading = twelfth_graders.groupby(["school_name"])['reading_score'].mean()
# Store the calculations in a dataframe
reading_scores_by_grade = pd.DataFrame({"9th": ninth_graders_reading, "10th": tenth_graders_reading, "11th": eleventh_graders_reading,
                                     "12th": twelfth_graders_reading})
reading_scores_by_grade.index.name = None
reading_scores_by_grade
```

## Scores by School Spending

```python
# Bins and Categories 
spending_bins = [0, 585, 630, 645, 680]
categories = ["$0 to 585", "$585-630", "$630-645", "$645-680"]
```

```python
per_school_summary.dtypes
```

```python
# Copy of per school dataframe
school_spending_df1 = per_school_summary.copy()
```

```python
# Formatting to remove the dollar sign from student budget so it can be changed to a float 
school_spending_df1["Per Student Budget"] = school_spending_df1["Per Student Budget"].replace({r'\$':''}, regex=True)
school_spending_df1["Per Student Budget"] = school_spending_df1["Per Student Budget"].astype("float64")
school_spending_df1.head()
```

```python
# Cutting the data and creating a new variable that gives spending ranges per student
school_spending_df1["Spending Ranges (Per Student)"] = pd.cut(school_spending_df1["Per Student Budget"], spending_bins,
                                                             labels=categories, include_lowest=True)
school_spending_df1.head()
```

```python
# Calculating the averages for each column to be stored in a new dataframe grouped by spending ranges
spending_math_scores = school_spending_df1.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df1.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df1.groupby(["Spending Ranges (Per Student)"])["Passing Math %"].mean()
spending_passing_reading = school_spending_df1.groupby(["Spending Ranges (Per Student)"])["Passing Reading %"].mean()
overall_passing_spending = school_spending_df1.groupby(["Spending Ranges (Per Student)"])["Passing Overall %"].mean()
```

```python
# Creating a new dataframe that shows average and passing scores by school spending per student
spending_summary = pd.DataFrame({"Average Math Score": spending_math_scores, "Average Reading Score": spending_reading_scores,
                                 "Passing Math %": spending_passing_math, "Passing Reading %": spending_passing_reading,
                                 "Passing Overall %": overall_passing_spending})
spending_summary
```

## Scores by School Size

```python
# Bins and Categories
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (0 to 1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Cutting the Students column and creating a new column that has different categories of school size
per_school_summary["School Size"] = pd.cut(per_school_summary['Total Students'], size_bins,
                                           labels=labels, include_lowest=True)
per_school_summary.head()
```

```python
# Calculating the averages for each column to be stored in a dataframe for school size
size_math_scores = per_school_summary.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["Passing Math %"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["Passing Reading %"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["Passing Overall %"].mean()
```

```python
# New DataFrame that stores average and passing scores for each range of school size
size_summary = pd.DataFrame({"Average Math Score": size_math_scores, "Average Reading Score": size_reading_scores,
                             "Passing Math %": size_passing_math, "Passing Reading %": size_passing_reading,
                             "Passing Overall %": size_overall_passing})
size_summary
```
